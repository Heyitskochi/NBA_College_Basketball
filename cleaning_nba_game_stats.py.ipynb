```python
import pandas as pd
```

```python
data= pd.read_csv("nba_games_stats.csv")
data
```

```python
cleandata=data[["Team", "WINorLOSS", "Date"]]
cleandata
```

```python
updated_df=cleandata.rename(columns={"WINorLOSS":"W_L"})
updated_df
```

```python
updated_df[['Year', 'Month', 'Day']] =updated_df['Date'].str.split('-', expand=True)
updated_df
```

```python
new_df=updated_df[['Team', 'W_L', 'Year']]
new_df
```

```python
new_df.to_csv(r'clean_nba_games_stats.csv', index=False, header=True)
```
